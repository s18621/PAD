```python
PATH ='./drive/MyDrive/Dane/PAD_03_PD.csv'
```

Zadanie 1 i 2

```python
import pandas as pd
sf_sal = pd.read_csv(PATH, sep=';')
sf_sal["owned_goods"] = sf_sal["owns_car"] + sf_sal["owns_TV"] + sf_sal["owns_house"] + sf_sal["owns_Phone"]
sf_sal
```

Zadanie 1

```python
by_state = sf_sal.groupby("Country")["ID"].count()
print(by_state)
```

Zadanie 3

```python
by_gender = sf_sal.groupby("gender")["owned_goods"].mean().round(2)
print(by_gender)
```

Zadanie 4

```python
df = pd.DataFrame(sf_sal).groupby("Country").agg({'owned_goods' : ['mean'], 'Age' : 'min'}).round(2)
df
```

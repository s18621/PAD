```python
from dateutil import parser

def str_to_date (date_time_str):
    #date_time_obj = datetime.strptime(date_time_str, '%Y/%m/%d')
    date_time_obj = parser.parse(date_time_str)
    return(date_time_obj)
```

```python
import numpy as np
import pandas as pd
import datetime
from datetime import datetime

sf_to = pd.read_csv(ORDERS, sep=',')
df_to = pd.DataFrame(sf_to)
df_to['order_date'] = df_to['order_date'].apply(str_to_date);
df_to
```

```python
value = df_to['tshirt_category'].describe()
print(value)
df_to['tshirt_category'].unique()
```

```python
df_to['tshirt_category'] = df_to['tshirt_category'].apply(lambda x: x.replace('Wh ','White '))
df_to['tshirt_category'] = df_to['tshirt_category'].apply(lambda x: x.replace('Bl ','Black '))
df_to['tshirt_category'] = df_to['tshirt_category'].apply(lambda x: x.replace('Tshirt ','T-Shirt '))
df_to.groupby('tshirt_category')['tshirt_category'].count()
```

```python
def type(val):
    val_split = val.split()
    if len(val_split) == 2: return val 
    elif len(val_split) == 1: return val_split[0]
    else: return val_split[1]

def colour(val):
    val_split = val.split()
    if len(val_split) == 3: return val_split[0]
    else: return None

def size(val):
    val_split = val.split()
    if len(val_split) == 3: return val_split[2]
    else: return None

df_to['tshirt_type']=df_to['tshirt_category'].apply(lambda x: type(x))
df_to['tshirt_size']=df_to['tshirt_category'].apply(lambda x: size(x))
df_to['tshirt_colour']=df_to['tshirt_category'].apply(lambda x: colour(x))
df_to
```
